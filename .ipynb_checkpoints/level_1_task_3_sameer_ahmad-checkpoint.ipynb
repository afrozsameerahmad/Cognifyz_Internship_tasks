```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_Dataset.csv')
```

```python
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)
```

```python
marker_cluster = MarkerCluster().add_to(m)
```

```python
def color_by_rating(rating):
    if rating >= 4.5:
        return 'green'
    elif rating >= 3.0:
        return 'orange'
    else:
        return 'red'
```

```python
for i, row in df.iterrows():
    popup_text = f"""
    <strong>Name:</strong> {row['Restaurant Name']}<br>
    <strong>City:</strong> {row['City']}<br>
    <strong>Rating:</strong> {row['Aggregate rating']}<br>
    <strong>Cuisines:</strong> {row['Cuisines']}<br>
    <strong>Votes:</strong> {row['Votes']}
    """
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=popup_text,
        icon=folium.Icon(color=color_by_rating(row['Aggregate rating']))
    ).add_to(marker_cluster)
```

```python
m.save("restaurant_map.html")
```

```python
country_counts = df['Country Code'].value_counts().reset_index()
country_counts.columns = ['Country_Code', 'Count']
print(country_counts)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=country_counts, y='Country_Code', x='Count', color='skyblue', ci=None)
for container in ax.containers:
    
    ax.bar_label(container,  fmt='%.0f',label_type='edge', padding=2)

plt.title('Restaurants by Country Code')
plt.xlabel('Number of Restaurants')
plt.ylabel('Country Code')
plt.tight_layout()
plt.show()
```

```python
city_counts = df['City'].value_counts().reset_index()
city_counts.columns = ['City', 'Count']
top_10_city = city_counts.sort_values(by='Count', ascending=False).head(10)
print(top_10_city)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=top_10_city, y='City', x='Count', color='skyblue', ci=None)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=2)
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Latitude', y='Aggregate rating', data=df, alpha=0.7, color='green')
plt.title('Latitude vs Aggregate Rating', fontsize=16)
plt.xlabel('Latitude', fontsize=14)
plt.ylabel('Aggregate Rating', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Aggregate rating', data=df, alpha=0.7, color='blue')
plt.title('Longitude vs Aggregate Rating', fontsize=16)
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Aggregate Rating', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix (Location vs Rating)', fontsize=16)
plt.tight_layout()
plt.show()
```
