```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_Dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
num_col = [
    "Longitude",
    "Latitude",
    "Average Cost for two",
    "Price range",
    "Aggregate rating",
    "Votes"
]
mean_values = df[num_col].mean()
median_values = df[num_col].median()
std_values = df[num_col].std()
print(" Mean Values \n ",mean_values)
print(" Median Values \n",median_values)
print(" Standard deviation Values \n",std_values)
```

```python
country_counts = df['Country Code'].value_counts().reset_index()
country_counts.columns = ['Country_Code', 'Count']
print(country_counts)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=country_counts, y='Country_Code', x='Count', color='skyblue', ci=None)
for container in ax.containers:
    
    ax.bar_label(container,  fmt='%.0f',label_type='edge', padding=2)

plt.title('Restaurants by Country Code')
plt.xlabel('Number of Restaurants')
plt.ylabel('Country Code')
plt.tight_layout()
plt.show()
```

```python
city_counts = df['City'].value_counts().reset_index()
city_counts.columns = ['City', 'Count']
top_10_city = city_counts.sort_values(by='Count', ascending=False).head(10)
print(top_10_city)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=top_10_city, y='City', x='Count', color='skyblue', ci=None)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=2)
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
cuisine_counts = df['Cuisines'].value_counts().reset_index()
cuisine_counts.columns = ['Cuisines', 'Count']
top_10_cuisines = cuisine_counts.sort_values(by='Count', ascending=False).head(10)
print(top_10_cuisines)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=top_10_cuisines, y='Cuisines', x='Count', color='skyblue', ci=None)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge', padding=2)
plt.title('Top 10 Cuisines by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisines')
plt.tight_layout()
plt.show()
```
