```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_Dataset.csv')
```

```python
#Histogram 
plt.figure(figsize=(8, 5))
plt.hist(df['Aggregate rating'], bins=10, color='lightblue', edgecolor='black')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# Plotting the bar plot
plt.figure(figsize=(12, 5))
sns.countplot(x='Aggregate rating', data=df, palette='Set2')
plt.title('Count of Each Rating')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
#Average rating by cuisine
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(8, 5))
cuisine_ratings.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Average Ratings by Cities
city_ratings = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(8, 5))
city_ratings.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Cities by Average Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Price Range vs Aggregate Rating 
plt.figure(figsize=(6, 4))
sns.boxplot(x="Price range", y="Aggregate rating", data=df)
plt.title("Price Range vs Aggregate Rating")
plt.show()
```

```python
#  Votes vs Aggregate Rating 
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Votes", y="Aggregate rating", data=df)
plt.title("Votes vs Aggregate Rating")
plt.show()
```

```python
# Online Delivery vs Aggregate Rating (Bar plot)
plt.figure(figsize=(6, 4))
sns.barplot(x="Has Online delivery", y="Aggregate rating", data=df)
plt.title("Online Delivery vs Aggregate Rating")
plt.show()
```

```python
# Table Booking vs Aggregate Rating (Bar plot)
plt.figure(figsize=(6, 4))
sns.barplot(x="Has Table booking", y="Aggregate rating", data=df)
plt.title("Table Booking vs Aggregate Rating")
plt.show()
```
